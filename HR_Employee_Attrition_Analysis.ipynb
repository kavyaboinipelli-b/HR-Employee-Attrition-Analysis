```python
import pandas as pd

df = pd.read_csv(r"C:\Users\91701\OneDrive\Desktop\HR-Employee-Attrition-Analysis\Dataset\WA_Fn-UseC_-HR-Employee-Attrition.csv")

print(df.shape)
df.head()
```

```python
import matplotlib.pyplot as plt

df['Attrition'].value_counts().plot(kind='bar')

plt.title('Employee Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Employees')

plt.show()
```

```python
print(df.columns.tolist())
```

```python
import matplotlib.pyplot as plt

dept_attrition = df.groupby('Department')['Attrition'].apply(lambda x: (x=='Yes').sum())

print(dept_attrition)

dept_attrition.plot(kind='bar')

plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Count')

plt.show()
```

```python
jobrole_attrition = df.groupby('JobRole')['Attrition'].apply(lambda x: (x=='Yes').sum())

jobrole_attrition.plot(kind='bar')

plt.title('Attrition by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Attrition Count')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
df['Attrition'].value_counts().plot(kind='bar')

plt.title('Employee Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Employees')

plt.tight_layout()

plt.show()
```

```python
salary_attrition = df.groupby('Attrition')['MonthlyIncome'].mean()

salary_attrition.plot(kind='bar')

plt.title('Average Salary by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Average Monthly Income')

plt.tight_layout()

plt.show()
```
